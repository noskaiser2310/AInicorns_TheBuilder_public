```python
# export_jsonl.py
import os
import json
from datasets import load_dataset

# tắt cảnh báo symlink (chỉ ẩn cảnh báo)
import os as _os
_os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

repo = "hchautran/vietnamese-wiki"  # adjust if cần
out_file = os.path.join(os.getcwd(), "vietnamese_wiki.jsonl")

# streaming=True để tiết kiệm RAM
ds = load_dataset(repo, split="train", streaming=True)  

count = 0
with open(out_file, "w", encoding="utf-8") as f:
    for example in ds:
        # example có thể chứa keys: 'id', 'title', 'clean_text'
        json_line = json.dumps(example, ensure_ascii=False)
        f.write(json_line + "\n")
        count += 1
        if count % 100000 == 0:
            print("Written", count, "examples")

print(f"Finished — wrote {count} examples to {out_file}")
```
